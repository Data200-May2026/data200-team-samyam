```python
from scipy.stats import f_oneway

low = df[df["AI Impact Level"]=="Low"]["Median Salary (USD)"]

medium = df[df["AI Impact Level"]=="Moderate"]["Median Salary (USD)"]

high = df[df["AI Impact Level"]=="High"]["Median Salary (USD)"]

f,p = f_oneway(low,medium,high)

print(f)
print(p)
```

```python
from scipy.stats import ttest_ind

inc = df[df["Job Status"]=="Increasing"]["Median Salary (USD)"]

dec = df[df["Job Status"]=="Declining"]["Median Salary (USD)"]

t,p = ttest_ind(inc,dec)

print(t)
print(p)
```
